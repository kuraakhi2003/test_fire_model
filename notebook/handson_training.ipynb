```python
X=df.drop("FWI",axis=1)
```

```python
y=df["FWI"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
def correlation(dataset,threshold):
    col_corr=set()
    col_matrix=dataset.corr()
    for i in range(len(col_matrix.columns)):
        for j in range(i):
            if abs(col_matrix.iloc[i,j])>threshold:
                col_corr.add(col_matrix.columns[i])
    return col_corr
        
    
```

```python
correl_columns=correlation(X_train,0.85)
```

```python
X_train.drop(columns=correl_columns,axis=1,inplace=True)
X_test.drop(columns=correl_columns,axis=1,inplace=True)
```

```python
X_test.drop(columns=correl_columns,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

```python
## Feature scaling or standardization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scale=scaler.fit_transform(X_train)
X_test_scale=scaler.transform(X_test)
```

```python
X_train_scale
```

```python
plt.subplots(figsize=(15,5))
plt.subplot(1,2,1)
sns.boxplot(data=X_train)
plt.title("Before scaling")
plt.subplot(1,2,2)
sns.boxplot(data=X_train_scale)
plt.title("After scaling")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
reg=LinearRegression()
reg.fit(X_train_scale,y_train)
ypred=reg.predict(X_test_scale)
mse=mean_squared_error(y_test,ypred)
score=r2_score(y_test,ypred)
print("Mean Squared Error:",mse)
print("R2 Score:",score)
plt.scatter(y_test,ypred)
```

## Lasso regresion

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error,r2_score
lassoreg=Lasso()
lassoreg.fit(X_train_scale,y_train)
ypred=lassoreg.predict(X_test_scale)
mse=mean_squared_error(y_test,ypred)
score=r2_score(y_test,ypred)
print("Mean Squared Error:",mse)
print("R2 Score:",score)
plt.scatter(y_test,ypred)
```

## cross validation lasso


```python
from sklearn.linear_model import LassoCV
lassocv=LassoCV(cv=5)
lassocv.fit(X_train_scale,y_train)
```

```python
lassocv.alphas_
```

```python
lassocv.mse_path_
```

```python
ypred=lassocv.predict(X_test_scale)
mse=mean_squared_error(y_test,ypred)
score=r2_score(y_test,ypred)
print("Mean Squared Error:",mse)
print("R2 Score:",score)
plt.scatter(y_test,ypred)
```

## ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error,r2_score
ridgereg=Ridge()
ridgereg.fit(X_train_scale,y_train)
ypred=ridgereg.predict(X_test_scale)
mse=mean_squared_error(y_test,ypred)
score=r2_score(y_test,ypred)
print("Mean Squared Error:",mse)
print("R2 Score:",score)
plt.scatter(y_test,ypred)
```

```python
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error,r2_score
ridgeregcv=RidgeCV(cv=5)
ridgeregcv.fit(X_train_scale,y_train)
ypred=ridgeregcv.predict(X_test_scale)
mse=mean_squared_error(y_test,ypred)
score=r2_score(y_test,ypred)
print("Mean Squared Error:",mse)
print("R2 Score:",score)
plt.scatter(y_test,ypred)
```
